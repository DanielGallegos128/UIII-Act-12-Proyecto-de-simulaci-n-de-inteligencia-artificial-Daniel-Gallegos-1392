```python
class Supervisor:
    # Constructor con los atributos solicitados
    def __init__(self, nombre, linea, empleados, incidencias, turno, calidad, planta):
        self.nombre = nombre
        self.linea = linea
        self.empleados = int(empleados)
        self.incidencias = int(incidencias)
        self.turno = turno
        self.calidad = float(calidad)  # Porcentaje de calidad (ej. 95.5)
        self.planta = planta

    # Función IA: Simula un diagnóstico inteligente basado en los atributos
    def simular_ia_diagnostico(self):
        if self.calidad >= 90 and self.incidencias <= 2:
            return "🤖 IA: Rendimiento Óptimo. Felicidades, mantiene la línea estable."
        elif self.calidad < 90 and self.incidencias > 2:
            return "🤖 IA: ALERTA. Alta incidencia y baja calidad. Se sugiere auditoría de procesos."
        else:
            return "🤖 IA: Rendimiento Regular. Monitorear personal en el próximo turno."


# Contenedor para nuestra lista de supervisores
lista_supervisores = []

# --- FUNCIONES DEL SISTEMA ---

def agregar_supervisor():
    print("\n--- 📝 REGISTRAR NUEVO SUPERVISOR ---")
    nombre = input("Nombre del Supervisor: ")
    linea = input("Línea de producción: ")
    empleados = input("Cantidad de empleados bajo su mando: ")
    incidencias = input("Reporte de incidencias (número de fallas): ")
    turno = input("Turno de trabajo (Matutino/Vespertino/Nocturno): ")
    calidad = input("Indicador de calidad (% de 0 a 100): ")
    planta = input("Planta / Ubicación: ")

    # Creamos el objeto y lo guardamos
    nuevo_super = Supervisor(nombre, linea, empleados, incidencias, turno, calidad, planta)
    lista_supervisores.append(nuevo_super)
    print(f"\n✅ ¡Supervisor {nombre} agregado con éxito al sistema!")

def listar_supervisores():
    print("\n--- 📊 LISTADO GENERAL DE SUPERVISORES Y DIAGNÓSTICO IA ---")
    if not lista_supervisores:
        print("No hay supervisores registrados en esta planta.")
        return

    # Cabecera de la tabla bien alineada
    # Usamos formato de strings con espacios fijos para que se vea como tabla
    print(f"{'ID':<4} | {'SUPERVISOR':<15} | {'LÍNEA':<10} | {'EMPLEADOS':<10} | {'INCIDENCIAS':<12} | {'TURNO':<10} | {'CALIDAD':<9} | {'PLANTA':<10}")
    print("-" * 95)

    # Imprimir filas
    for indice, superv in enumerate(lista_supervisores):
        print(f"{indice:<4} | {superv.nombre:<15} | {superv.linea:<10} | {superv.empleados:<10} | {superv.incidencias:<12} | {superv.turno:<10} | {superv.calidad:<8}% | {superv.planta:<10}")
        # Mostramos la simulación de IA justo abajo de su registro
        print(f"     └─ {superv.simular_ia_diagnostico()}")
        print("-" * 95)

def actualizar_supervisor():
    print("\n--- 🔄 ACTUALIZAR DATOS DE SUPERVISOR ---")
    if not lista_supervisores:
        print("No hay datos para actualizar.")
        return

    # Mostramos los existentes para que elijan el ID
    listar_supervisores()
    id_modificar = int(input("\nIngrese el ID del supervisor que desea modificar: "))

    # Accedemos al objeto seleccionado
    superv = lista_supervisores[id_modificar]
    print(f"\nModificando a: {superv.nombre}. Deja en blanco si no deseas cambiar el dato.")

    # Validaciones simples de entrada
    nuevo_nombre = input(f"Nombre actual ({superv.nombre}): ")
    if nuevo_nombre != "": superv.nombre = nuevo_nombre

    nueva_linea = input(f"Línea actual ({superv.linea}): ")
    if nueva_linea != "": superv.linea = nueva_linea

    nuevos_emp = input(f"Empleados actuales ({superv.empleados}): ")
    if nuevos_emp != "": superv.empleados = int(nuevos_emp)

    nuevas_inc = input(f"Incidencias actuales ({superv.incidencias}): ")
    if nuevas_inc != "": superv.incidencias = int(nuevas_inc)

    nuevo_turno = input(f"Turno actual ({superv.turno}): ")
    if nuevo_turno != "": superv.turno = nuevo_turno

    nueva_cal = input(f"Calidad actual ({superv.calidad}%): ")
    if nueva_cal != "": superv.calidad = float(nueva_cal)

    nueva_planta = input(f"Planta actual ({superv.planta}): ")
    if nueva_planta != "": superv.planta = nueva_planta

    print("\n✅ ¡Datos actualizados y re-analizados por la IA!")

def borrar_supervisor():
    print("\n--- ❌ BORRAR SUPERVISOR DEL SISTEMA ---")
    if not lista_supervisores:
        print("No hay supervisores para eliminar.")
        return

    listar_supervisores()
    id_borrar = int(input("\nIngrese el ID del supervisor que desea eliminar: "))

    # Eliminamos de la lista y guardamos el nombre para avisar
    eliminado = lista_supervisores.pop(id_borrar)
    print(f"\n🗑️ El supervisor {eliminado.nombre} ha sido removido del sistema.")

while True:
    print("\n=============================================")
    print("🤖 SISTEMA IA: GESTIÓN DE PLANTAS INDUX 🤖")
    print("=============================================")
    print("1. Registrar Supervisor")
    print("2. Ver Planta y Diagnóstico IA (Tabla)")
    print("3. Actualizar Datos de Supervisor")
    print("4. Dar de Baja Supervisor")
    print("5. Salir del Sistema")
    print("=============================================")

    opcion = input("Selecciona una opción (1-5): ")

    if opcion == "1":
        agregar_supervisor()
    elif opcion == "2":
        listar_supervisores()
    elif opcion == "3":
        actualizar_supervisor()
    elif opcion == "4":
        borrar_supervisor()
    elif opcion == "5":
        print("\n👋 Saliendo del asistente de IA. ¡Buen día en la planta!")
        break
    else:
        print("\n❌ Opción no válida. Intenta de nuevo.")
```
